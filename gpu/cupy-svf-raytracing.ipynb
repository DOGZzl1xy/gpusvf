```python
import os, os.path
import rasterio as rio
from osgeo import gdal
from osgeo.gdalconst import *
from matplotlib import pyplot as plt
import numpy as np
import rasterio
import time
import math
import struct


dsmfile = demfile = r"E:\Code\gpusvf\gpusvf\test_data\BG_1F.tif"
treefile = r"E:\Code\gpusvf\gpusvf\test_data\Vege_1F.tif"


with rasterio.open(dsmfile) as dsmlayer:
    dsmimg = dsmlayer.read(1)
with rasterio.open(treefile) as treelayer:
    treeimg = treelayer.read(1)

gdal_dsm = gdal.Open(dsmfile)
dsm = gdal_dsm.ReadAsArray().astype(float)
tree = treeimg.astype(float)
geotransform = gdal_dsm.GetGeoTransform()
scale = 1 / geotransform[1]

# 处理 NoData 值
nodata_value = -3.4028234663852886e+38
dsm = np.where(dsm == nodata_value, np.nan, dsm)
tree = np.where(tree < 0, 0, tree)

# 用邻近值插值 NaN 值，或者简单地设置为平均高程
if np.isnan(dsm).any():
    valid_data = dsm[~np.isnan(dsm)]
    if len(valid_data) > 0:
        min_valid = np.min(valid_data)
        dsm = np.where(np.isnan(dsm), min_valid, dsm)

if np.isnan(tree).any():
    valid_data = tree[~np.isnan(tree)]
    if len(valid_data) > 0:
        min_valid = np.min(valid_data)
        tree = np.where(np.isnan(tree), min_valid, tree)

print(f"DSM shape: {dsm.shape}, Tree shape: {tree.shape}")
print(f"DSM value range: {np.min(dsm)} to {np.max(dsm)}, Tree value range: {np.min(tree)} to {np.max(tree)}")
print(f"GeoTransform: {geotransform}")
print(f"Scale: {scale}")
```

```python
# Block 2: GPU编码文件（CuPy内核）
import cupy as cp

svf_kernel_code = r'''
#define PI 3.1415926f
#define NO_DATA_VALUE -9999.0f

extern "C" __global__ void svfcalculator(
    float* lattice_out, float* lattice, float* tree, float scale, int imageW, int imageH) {
    int x = threadIdx.x + blockIdx.x * blockDim.x;
    int y = threadIdx.y + blockIdx.y * blockDim.y;

    if (x >= imageW || y >= imageH) return;

    int index4 = x + y * imageW;
    float height0 = lattice[index4];
    float tree0 = tree[index4];

    // 如果当前位置有树且树高>0，认为站在树下，SVF=0
    if (tree0 > 0.0f) {
        lattice_out[index4] = 0.0f;
        return;
    }

    // 检查 NoData 值
    if (height0 < -1000 || isnan(height0)) {
        lattice_out[index4] = 1.0;
        return;
    }

    float SVF_res = 0.0;

    for (int angle = 0; angle < 360; angle++) {
        float betaMax = 0.0;
        float alpha = angle * PI / 180.0;
        bool blocked = false;

        for (int dist = 1; dist < 200; dist++) {
            float dx = dist * cos(alpha);
            float dy = dist * sin(alpha);

            float x1_f = x + dx;
            float y1_f = y + dy;

            int x1 = round(x1_f);
            int y1 = round(y1_f);

            if (x1 < 0 || x1 >= imageW || y1 < 0 || y1 >= imageH) continue;

            int index2 = x1 + y1 * imageW;
            float height1 = lattice[index2];
            float tree1 = tree[index2];

            // 如果遇到树，则高度为地面高程+6米，并视为完全遮挡，直接break
            if (tree1 > 0.0f) {
                height1 = lattice[index2] + 6.0f;
                float distance = sqrtf(dx*dx + dy*dy) * scale;
                if (distance > 0.0f) {
                    float height_diff = height1 - height0;
                    float beta = atanf(height_diff / distance);
                    if (beta > betaMax) betaMax = beta;
                }
                blocked = true;
                break;
            }

            // 如果遇到比当前点高的建筑物，也视为完全遮挡，直接break
            if (height1 > height0 + 0.5f) { // 0.5f为容差
                float distance = sqrtf(dx*dx + dy*dy) * scale;
                if (distance > 0.0f) {
                    float height_diff = height1 - height0;
                    float beta = atanf(height_diff / distance);
                    if (beta > betaMax) betaMax = beta;
                }
                blocked = true;
                break;
            }

            if (height1 < -1000 || isnan(height1)) continue;
        }

        float cos_beta = cosf(betaMax);
        if (cos_beta > 0.0f) {
            SVF_res += cos_beta * cos_beta;
        }
    }

    float result = SVF_res / 360.0f;
    if (result < 0.0f) result = 0.0f;
    if (result > 1.0f) result = 1.0f;
    lattice_out[index4] = result;
}
'''

# 编译内核
svf_kernel = cp.RawKernel(svf_kernel_code, 'svfcalculator')
```

```python
# Block 3: 剩下的其他部分
import time

def svfCalculator_RayTracingOnGPU(dsm, tree, scale):
    '''This code is used to calculate the sky view factor using the ray-tracing
    algorithm based on the GPU acceleration
    last modified Jan 27, 2021
    by Xiaojiang Li, Temple University
    
    Parameters:
        dsm: the numpy array of the digital surface model
        scale: is the scale of the image, read from the gdal,
                1px of 2 feet, scale is 0.5; 1px of 3 feet, scale is 0.3333
    '''
    
    # 将输入数据转换为CuPy数组
    px = cp.asarray(dsm).astype(cp.float32)
    tree_px = cp.asarray(tree).astype(cp.float32)
    
    height, width = px.shape
    nb_pixels = height * width
    
    # 设置线程块和网格大小
    nb_ThreadsX = 8
    nb_ThreadsY = 8
    nb_blocksX = (width // nb_ThreadsX) + 1
    nb_blocksY = (height // nb_ThreadsY) + 1

    tps1 = time.time()
    
    # 创建输出数组
    newLattice_gpu = cp.empty_like(px)
    
    # 调用内核
    svf_kernel(
        (nb_blocksY, nb_blocksX),
        (nb_ThreadsY, nb_ThreadsX),
        (newLattice_gpu, px, tree_px, cp.float32(scale), width, height)
    )
    
    # 等待GPU计算完成
    cp.cuda.Stream.null.synchronize()
    
    # 将结果复制回主机内存
    bwPx = cp.asnumpy(newLattice_gpu)
    
    return bwPx

# Block 4: 保存为GeoTIFF的函数
def saverasternd(gdal_data, filename, raster):
    rows = gdal_data.RasterYSize
    cols = gdal_data.RasterXSize
    
    outDs = gdal.GetDriverByName("GTiff").Create(filename, cols, rows, int(1), GDT_Float32)
    outBand = outDs.GetRasterBand(1)
    
    # write the data
    outBand.WriteArray(raster, 0, 0)
    # flush data to disk, set the NoData value and calculate stats
    outBand.FlushCache()
    outBand.SetNoDataValue(-9999)
    
    # georeference the image and set the projection
    outDs.SetGeoTransform(gdal_data.GetGeoTransform())
    outDs.SetProjection(gdal_data.GetProjection())
    
    # close the datasets
    outBand = None
    outDs = None
    gdal_data = None
```

```python
# Block 5: 主处理流程 (根据您的环境修改)
# 使用您已经加载的数据进行处理
t1 = time.time()

# 使用您已经读取的dsm数据
dsm_img = dsm
tree_img = tree

print(f"Processing image with shape: {dsm_img.shape}")

# 计算SVF
res = svfCalculator_RayTracingOnGPU(dsm_img, tree_img, np.float32(scale))

# 处理无穷大和NaN值
res = np.where(np.isinf(res), 1.0, res)  # 将无穷大值设为1
res = np.where(np.isnan(res), 1.0, res)  # 将NaN值设为1
res = np.clip(res, 0.0, 1.0)

t2 = time.time()
print(f'The time elapsed is: {t2-t1:.2f} seconds')

# 保存结果
output_filename = r"E:\Code\gpusvf\gpusvf\test_data\svf_BG_1F.tif"
saverasternd(gdal_dsm, output_filename, np.asarray(res))
print(f"Saved result to {output_filename}")

# 可视化结果
plt.imshow(np.asarray(res))
plt.colorbar()
plt.title("Sky View Factor")
plt.savefig(r"E:\Code\gpusvf\gpusvf\test_data\svf_BG_1F.png", dpi=300)
plt.show()

# 清理资源
gdal_dsm = None
```
